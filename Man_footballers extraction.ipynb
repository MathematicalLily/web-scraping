```python
import pandas as pd
from bs4 import BeautifulSoup
import requests as r
```

```python
response = r.get("https://en.wikipedia.org/wiki/List_of_Manchester_City_F.C._players")
wiki_text = response.text
```

```python
soup = BeautifulSoup(wiki_text, 'html.parser')
print(soup)
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
print(soup.a)
```

```python
my_table = soup.find('table',{'class':'wikitable sortable'})
print(my_table)
```

```python
links = my_table.findAll('a')
print(links)

```

```python
Title = []
for link in links:
    Title.append(link.get('title'))
    print(Title)
    
   
```

```python
import pandas as pd
df = pd.DataFrame()
df ['Title'] = Title
df
```

```python
pd = df.to_csv('Manchester footballers')
```
